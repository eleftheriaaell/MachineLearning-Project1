# 3. Gradient Descent

First function with parametres given in the project:

**x1 = 0, x2 = 0, learning rate = 0.5, iterations = 10**

```python
#gradient descent for f1, with default parametres

import numpy as np
import matplotlib.pyplot as plt

x1 = 0
x2 = 0
learning_rate = 0.5
max_iterations = 10

func1 = np.array([])

for i in range(max_iterations):
  
  #calculating derivatives for each variable
  x1_derivative = 2 * (x1 - 2)
  x2_derivative = 2 * (x2 - 3)

  #updating rules for each variable
  x1 = x1 - learning_rate * x1_derivative
  x2 = x2 - learning_rate * x2_derivative

  func1 = np.append(func1, pow((x1 - 2), 2) + pow((x2 - 3), 2))
  print(f"Iteration {i}: [x1,x2] = {x1, x2}, f1(x1,x2) = {func1[i]}")

#plotting
plt.plot(func1)
plt.xlim(0, max_iterations-1)

plt.xlabel("Iteration Number")
plt.ylabel("f1(x1,x2)")

plt.show()
```

**Παρατηρήσεις για την πρώτη συνάρτηση:**

Βλέπουμε ότι με σταθερές τιμές αρχικοποίησης x1 = 0, x2 = 0, για μικρές τιμές learning rate απαιτούνται περισσότερες επαναλήψεις για να συγκλίνουμε σε στάσιμο σημείο της συνάρτησης. Όσο αυξάνεται το learning rate παρατηρείτε σύγκλιση στο στάσιμο σημείο σε λιγότερες επαναλήψεις.

Μια άλλη παρατήρηση είναι ότι όταν το learning rate είναι πολύ κοντά στο 0 ή 1, χρειάζεται πολύ μεγάλος αριθμός επαναλήψεων μέχρι τη σύγλιση σε στάσιμο σημείο. Έγινε δοκιμή και με 200000 iterations ανεπυτηχώς. Έστω, πολύ κοντά στο 0: lr = 0.00000009 και πολύ κοντά στο 1: lr = 0.999999.

Επίσης για μικρό learning rate παρατηρούμε ότι για μικρότερες τιμές αρχικοποίησης χρειάζονται λιγότερες επαναλήψεις για να συγκλίνει σε σταθερό σημείο, ενώ για μεγαλύτερες τιμές χρειάζονται περισσότερες επαναλήψεις. Όσο μικρότερο το learning rate, τόσο περισσότερο μεγαλώνει η διαφορά στο πλήθος των επαναλήψεων που χρειάζονται για σύγκλιση, ανάμεσα σε μικρές και μεγάλες τιμές αρχικοποίησης.

Ενώ για μεγάλα learning rate παρατηρούμε ότι, είτε είναι μικρές οι τιμές αρχικοποίησης είτε μεγάλες, το πλήθος των επαναλήψεων μέχρι την σύγκλιση είναι πάνω κάτω το ίδιο.


Second function with parametres given in the project:

**x1 = 0, x2 = 0, learning rate = 0.5, iterations = 10**

```python
#gradient descent for f2, with default parametres

import numpy as np
import matplotlib.pyplot as plt

x1 = 0
x2 = 0
learning_rate = 0.5
max_iterations = 10

func2 = np.array([])

for i in range(max_iterations):
  
  #calculating derivatives for each variable
  x1_derivative = 40 * ((x1 + 3) - pow((x2 - 3), 2))
  x2_derivative = - 2 * (4 - x2) - 80 * ((x1 + 3) - pow((x2 - 3), 2) ) * (x2 - 3)

  #updating rules for each variable
  x1 = x1 - learning_rate * x1_derivative
  x2 = x2 - learning_rate * x2_derivative
  
  func2 = np.append(func2, pow((4 - x2), 2) + 20 * pow((x1 + 3) - pow((x2 - 3), 2), 2))
  print(f"Iteration {i}: [x1,x2] = {x1, x2}, f2(x1,x2) = {func2[i]}")

#plotting
plt.plot(func2)
plt.xlim(0, max_iterations-1)

plt.xlabel("Iteration Number")
plt.ylabel("f2(x1,x2)")

plt.show()
```
