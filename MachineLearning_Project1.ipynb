# 3. Gradient Descent

First function with parametres given in the project:

**x1 = 0, x2 = 0, learning rate = 0.5, iterations = 10**

```python
#gradient descent for f1, with default parametres

import numpy as np
import matplotlib.pyplot as plt

x1 = 0
x2 = 0
learning_rate = 0.5
max_iterations = 10

func1 = np.array([])

for i in range(max_iterations):
  
  #calculating derivatives for each variable
  x1_derivative = 2 * (x1 - 2)
  x2_derivative = 2 * (x2 - 3)

  #updating rules for each variable
  x1 = x1 - learning_rate * x1_derivative
  x2 = x2 - learning_rate * x2_derivative

  func1 = np.append(func1, pow((x1 - 2), 2) + pow((x2 - 3), 2))
  print(f"Iteration {i}: [x1,x2] = {x1, x2}, f1(x1,x2) = {func1[i]}")

#plotting
plt.plot(func1)
plt.xlim(0, max_iterations-1)

plt.xlabel("Iteration Number")
plt.ylabel("f1(x1,x2)")

plt.show()
```

```python
#gradient descent for f2, with default parametres

import numpy as np
import matplotlib.pyplot as plt

x1 = 0
x2 = 0
learning_rate = 0.5
max_iterations = 10

func2 = np.array([])

for i in range(max_iterations):
  
  #calculating derivatives for each variable
  x1_derivative = 40 * ((x1 + 3) - pow((x2 - 3), 2))
  x2_derivative = - 2 * (4 - x2) - 80 * ((x1 + 3) - pow((x2 - 3), 2) ) * (x2 - 3)

  #updating rules for each variable
  x1 = x1 - learning_rate * x1_derivative
  x2 = x2 - learning_rate * x2_derivative
  
  func2 = np.append(func2, pow((4 - x2), 2) + 20 * pow((x1 + 3) - pow((x2 - 3), 2), 2))
  print(f"Iteration {i}: [x1,x2] = {x1, x2}, f2(x1,x2) = {func2[i]}")

#plotting
plt.plot(func2)
plt.xlim(0, max_iterations-1)

plt.xlabel("Iteration Number")
plt.ylabel("f2(x1,x2)")

plt.show()
```
